```python
class CVBuilder:
    def __init__(self):
        self.name = ""
        self.email = ""
        self.address = ""
        self.education = {}
        self.work_experience = []
        self.skills = []
        self.languages = []

    def get_personal_details(self):
        self.name = input("Enter your full name: ")
        self.email = input("Enter your email address: ")
        self.address = input("Enter your address: ")

    def get_education_details(self):
        print("\nEnter your educational details:")
        for level in ["High School", "Secondary School", "Bachelors", "Masters"]:
            school_name = input(f"Enter the name of {level} institution: ")
            year = input(f"Enter the year of completion for {level}: ")
            self.education[level] = {"Institution": school_name, "Year": year}

    def get_work_experience(self):
        print("\nEnter your work experience (type 'done' when finished):")
        while True:
            company = input("Company Name: ")
            if company.lower() == "done":
                break
            role = input("Role/Position: ")
            duration = input("Duration (e.g., Jan 2020 - Dec 2021): ")
            self.work_experience.append({"Company": company, "Role": role, "Duration": duration})

    def get_skills(self):
        print("\nEnter your soft skills (type 'done' when finished):")
        while True:
            skill = input("Skill: ")
            if skill.lower() == "done":
                break
            self.skills.append(skill)

    def get_languages(self):
        print("\nEnter languages you know (type 'done' when finished):")
        while True:
            language = input("Language: ")
            if language.lower() == "done":
                break
            self.languages.append(language)

    def display_cv(self):
        print("\n--- CV ---\n")
        print(f"Name: {self.name}")
        print(f"Email: {self.email}")
        print(f"Address: {self.address}")
        print("\n--- Education ---")
        for level, details in self.education.items():
            print(f"{level}: {details['Institution']} ({details['Year']})")

        print("\n--- Work Experience ---")
        for exp in self.work_experience:
            print(f"{exp['Role']} at {exp['Company']} ({exp['Duration']})")

        print("\n--- Soft Skills ---")
        for skill in self.skills:
            print(f"- {skill}")

        print("\n--- Languages ---")
        for language in self.languages:
            print(f"- {language}")

    def build_cv(self):
        self.get_personal_details()
        self.get_education_details()
        self.get_work_experience()
        self.get_skills()
        self.get_languages()
        self.display_cv()


# Create CV
cv = CVBuilder()
cv.build_cv()

d
```
